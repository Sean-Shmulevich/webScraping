```python
sel = Selector(text = webData)
```

```python
#Creates array for all data on website and converts it to text
a = sel.css("*::text").getall()


links = []
for i in a:
    if(i[0:4] == 'http' and len(i) > 4 and not (i[len(i)-3:len(i)] == "png")):
        links.append(i)
```
